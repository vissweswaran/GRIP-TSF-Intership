### Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Read DataSet

```python
df = pd.read_csv("SampleSuperstore.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

##  1.REGION WISE PROFIT

```python
RP = df.groupby('Region',as_index=False)['Profit'].sum()
RP.plot(kind='bar',x='Region',y='Profit')
```

### "Central","In this region you have to take some action for improvement and better profit"

##  2.REGION WISE SALES

```python
RS = df.groupby('Region',as_index=False)['Sales'].sum()
RS.plot(kind='bar',x='Region',y='Sales')
```

### "South","In this region you have to take some Sales improvement for better Profit"

##  3.SHIP MODE OF PRODUCTS

```python
SS = df.groupby('Ship Mode',as_index=False)['Sales'].sum()
SS.plot(kind='line',x='Ship Mode',y='Sales')
```

### Shipment of Products in the Same Day OR first class for better results in Sales

## 4.SALES REPORT BY REGION-WISE

```python
sns.countplot(data=df,x="Region",hue="Segment")
```

### Improve of Sales for Home Office for better Profit

## 5.STATE-WISE SALES REPORT

```python
SS=df.groupby('State')['Sales'].sum()
SS.plot(kind='bar',figsize=(20,5),x="State")
```

### Additional Concentration for Sales in States are Lowa,Kansas,Maine,D.of.Columbia,New Mexico,North Dakota,South Dakota,West Virginia,Wyoming

## 6.STATE-WISE PROFIT REPORT

```python
SP=df.groupby('State',as_index=False)['Profit'].sum()
SP.plot(kind='bar',x='State',figsize=(20,5))
```

### Texas,Lllinois,Chio,Pennsylvania are major States which are giving loss compare with Sales Report make some ideas to bring back to Profit

## 7.CATEGORY-WISE PROFIT REPORT

```python
CR=df.groupby('Category',as_index=False)['Profit'].sum()
CR.plot(kind='bar',color='green',x="Category")
```

### Comparing with sales and Profit of Furniture sales is high but comparing with Other Categories which is very low in profit

## 8.SUB-CATEGORY-WISE REPORT

```python
sns.countplot(data=df,x="Category",hue="Sub-Category",palette="Set2")
```

```python
sns.catplot(x="Category", hue="Sub-Category", col="Region",
                data=df, kind="count",palette="Set2",height=4, aspect=.8)
```

### Improvement in Sales of Technology things to gain more profit
